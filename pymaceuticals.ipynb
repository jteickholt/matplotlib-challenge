### Analysis of potential treatments (Capomulin, Infubinol, Ketapril, and Placebo) of squamous cell carcinoma (SCC)

#### Import needed libraries

```python
# Define matplotlib inline so plots can be displayed in the notebook

%matplotlib inline

# Import libraries needed for the analysis

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

# Hide warning messages in notebook
# import warnings
# warnings.filterwarnings('ignore')
```

#### Read in data from csv files that are stored outside of repo. 

#### clinicaltrial_date.csv contains the measures over time for each mouse.  mouse_drug_data.csv contains the mouse id, along with drug assigned to the mouse.

```python
# Read the Mouse and Drug Data and the Clinical Trial Data

clinicaltrial_data_to_load=os.path.join('..','..', 'RU-HOU-DATA-PT-09-2019-U-C','Homework','05-Matplotlib','Instructions','Pymaceuticals','data','clinicaltrial_data.csv')
mousedrug_data_to_load=os.path.join('..','..', 'RU-HOU-DATA-PT-09-2019-U-C','Homework','05-Matplotlib','Instructions','Pymaceuticals','data','mouse_drug_data.csv')

clinicaltrial_df=pd.read_csv(clinicaltrial_data_to_load)
clinicaltrial_df.shape

mouse_df=pd.read_csv(mousedrug_data_to_load)
mouse_df.shape

# Combine the data into a single dataset
combined_df=pd.merge(clinicaltrial_df, mouse_df, how='left', on=['Mouse ID','Mouse ID'])


#Keep only the drugs were are interested in
combined_df=combined_df[combined_df['Drug'].isin(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'])]

combined_df.head()
```

### Examine Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

grp_drug_time_df=combined_df.groupby(['Drug','Timepoint'])

avg_tumvol=grp_drug_time_df['Tumor Volume (mm3)'].mean()
stderr_tumvol=grp_drug_time_df['Tumor Volume (mm3)'].sem()

#Unstack the grouped df to get the data in a format that can be used for plotting
#Reset index so Timepoint can be used in plot

grp_drug_unstack_df=avg_tumvol.unstack(level=0)
grp_drug_unstack_df=grp_drug_unstack_df.reset_index()
stderr_unstack_df=stderr_tumvol.unstack(level=0)

#Set up variables to hold mean and std error for each drug from Data Frames
Timepoint=grp_drug_unstack_df['Timepoint']
CapomulinVol=grp_drug_unstack_df['Capomulin']
InfubinolVol=grp_drug_unstack_df['Infubinol']
KetaprilVol=grp_drug_unstack_df['Ketapril']
PlaceboVol=grp_drug_unstack_df['Placebo']

CapomulinStdErr=stderr_unstack_df['Capomulin']
InfubinolStdErr=stderr_unstack_df['Infubinol']
KetaprilStdErr=stderr_unstack_df['Ketapril']
PlaceboStdErr=stderr_unstack_df['Placebo']

#Plot the means and standard erors on same plot
fig, ax = plt.subplots()

ax.errorbar(Timepoint, CapomulinVol, CapomulinStdErr, color='red', marker='o', label='Capomulin')
ax.errorbar(Timepoint, InfubinolVol, InfubinolStdErr, color='blue', marker='^', label='Infubinol')
ax.errorbar(Timepoint, KetaprilVol, KetaprilStdErr, color='green', marker='s',label='Ketapril')
ax.errorbar(Timepoint, PlaceboVol, PlaceboStdErr, color='black', marker='D', label='Placebo')

#Add a legend and label the axis
ax.legend(loc='best')
ax.set_xlabel('Time (Days)')
ax.set_ylabel('Tumor Volume (mm3)')
ax.set_title('Tumor Response to Treatment')
ax.grid(b=True, which='major', axis='y')

```

### Examine Metastatic Response to Treatment

```python
# Create the Chart of Metastatic Sites


#Calculate the averge number of metestatic sites by drug, timepoint
avg_metastatic_sites=grp_drug_time_df['Metastatic Sites'].mean()
stderr_metastatic_sites=grp_drug_time_df['Metastatic Sites'].sem()

#Unstack the series for each.  Reset the index for the first, so Timepoint can be used
avg_sites_unstack_df=avg_metastatic_sites.unstack(level=0)
avg_sites_unstack_df=avg_sites_unstack_df.reset_index()
stderr_unstack_df=stderr_metastatic_sites.unstack(level=0)

#Set up variables to hold mean and std error for each drug from Data Frames
Timepoint=avg_sites_unstack_df['Timepoint']
CapomulinSites=avg_sites_unstack_df['Capomulin']
InfubinolSites=avg_sites_unstack_df['Infubinol']
KetaprilSites=avg_sites_unstack_df['Ketapril']
PlaceboSites=avg_sites_unstack_df['Placebo']

CapomulinSitesStdErr=stderr_unstack_df['Capomulin']
InfubinolSitesStdErr=stderr_unstack_df['Infubinol']
KetaprilSitesStdErr=stderr_unstack_df['Ketapril']
PlaceboSitesStdErr=stderr_unstack_df['Placebo']

#Plot the means and standard erors on same plot
fig, ax = plt.subplots()

ax.errorbar(Timepoint, CapomulinSites, CapomulinSitesStdErr, color='red', marker='o', label='Capomulin')
ax.errorbar(Timepoint, InfubinolSites, InfubinolSitesStdErr, color='blue', marker='^', label='Infubinol')
ax.errorbar(Timepoint, KetaprilSites, KetaprilSitesStdErr, color='green', marker='s', label='Ketapril')
ax.errorbar(Timepoint, PlaceboSites, PlaceboSitesStdErr, color='black', marker='D', label='Placebo')


#Add a legend and label the axis
ax.legend(loc='best')
ax.set_xlabel('Treatment Duration (Days)')
ax.set_ylabel('Met. Sites')
ax.set_title('Metastatic Spread During Treatment')
ax.grid(b=True, which='major', axis='y')


```

### Survival Rates

```python
#Get a count of the number of Mice at each timepoint by drug
mouse_count=grp_drug_time_df['Mouse ID'].count()

#Unstack the series and restet index so Timepoint can be used
mouse_count
mouse_count_unstack_df=mouse_count.unstack(level=0)
mouse_count_unstack_df=mouse_count_unstack_df.reset_index()

#Here I get the number of mice in each drug that started at Timepoint 0
#Given that they all started at 25, I could have done this once, but doign for all would
#allow for the number of mice starting to be different by drug
#the df.loc will return a series with one element, so used .get(key=0) to extract the value 

CapomulinNum=mouse_count_unstack_df.loc[mouse_count_unstack_df['Timepoint']==0, 'Capomulin'].get(key=0)
InfubinolNum=mouse_count_unstack_df.loc[mouse_count_unstack_df['Timepoint']==0, 'Infubinol'].get(key=0)
KetaprilNum=mouse_count_unstack_df.loc[mouse_count_unstack_df['Timepoint']==0, 'Infubinol'].get(key=0)
PlaceboNum=mouse_count_unstack_df.loc[mouse_count_unstack_df['Timepoint']==0, 'Infubinol'].get(key=0)

#  I then add the number that started into the dataframe with the number still
#  alive at each timepoint.

mouse_count_unstack_df['Capomulin Mice']=CapomulinNum
mouse_count_unstack_df['Infubinol Mice']=InfubinolNum
mouse_count_unstack_df['Ketapril Mice']=KetaprilNum
mouse_count_unstack_df['Placebo Mice']=PlaceboNum

#  Caclculate the percentage of mice still alive at each time point
                       
Timepoint = mouse_count_unstack_df['Timepoint']
CapomulinPerc = (mouse_count_unstack_df['Capomulin']/mouse_count_unstack_df['Capomulin Mice'])*100
InfubinolPerc = (mouse_count_unstack_df['Infubinol']/mouse_count_unstack_df['Infubinol Mice'])*100
KetaprilPerc = (mouse_count_unstack_df['Ketapril']/mouse_count_unstack_df['Ketapril Mice'])*100
PlaceboPerc = (mouse_count_unstack_df['Placebo']/mouse_count_unstack_df['Placebo Mice'])*100

#Plot the means and standard erors on same plot
fig, ax = plt.subplots()

plt.plot(Timepoint, CapomulinPerc, color='red', marker='o', label='Capomulin')
plt.plot(Timepoint, InfubinolPerc, color='blue', marker='^', label='Infubinol')
plt.plot(Timepoint, KetaprilPerc, color='green', marker='s', label='Ketapril')
plt.plot(Timepoint, PlaceboPerc, color='black', marker='D', label='Placebo')


#Add a legend and label the axis
ax.legend(loc='best')
ax.set_xlabel('Time (Days)')
ax.set_ylabel('Survival Rate (%)')
ax.set_title('Survival During Treatment')
ax.grid(b=True, which='major', axis='both')

```

### Examine Change in Tumor Size During Study Period

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

grp_drug_time_df=combined_df.groupby(['Drug','Timepoint'])

avg_tumvol=grp_drug_time_df['Tumor Volume (mm3)'].mean()

#Unstack the grouped df to get the data in a format that can be used for plotting
#Reset index so Timepoint can be used in plot

grp_drug_unstack_df=avg_tumvol.unstack(level=0)
grp_drug_unstack_df=grp_drug_unstack_df.reset_index()

#Here I get the average starting tumor volume (timepoint=0) and the average final (timepoint=45)
#I then calculate the average change for each drug

Capvolstart = grp_drug_unstack_df.loc[grp_drug_unstack_df['Timepoint']==0, 'Capomulin'].get(key=0)
Capvolend =   grp_drug_unstack_df.loc[grp_drug_unstack_df['Timepoint']==45, 'Capomulin'].reset_index(drop='True').get(key=0)
Capvolchange = ((Capvolend - Capvolstart)/Capvolstart)*100

Infvolstart = grp_drug_unstack_df.loc[grp_drug_unstack_df['Timepoint']==0, 'Infubinol'].get(key=0)
Infvolend =   grp_drug_unstack_df.loc[grp_drug_unstack_df['Timepoint']==45, 'Infubinol'].reset_index(drop='True').get(key=0)
Infvolchange = ((Infvolend - Infvolstart)/Infvolstart)*100

Ketvolstart = grp_drug_unstack_df.loc[grp_drug_unstack_df['Timepoint']==0, 'Ketapril'].get(key=0)
Ketvolend =   grp_drug_unstack_df.loc[grp_drug_unstack_df['Timepoint']==45, 'Ketapril'].reset_index(drop='True').get(key=0)
Ketvolchange = ((Ketvolend - Ketvolstart)/Ketvolstart)*100

Plavolstart = grp_drug_unstack_df.loc[grp_drug_unstack_df['Timepoint']==0, 'Placebo'].get(key=0)
Plavolend =   grp_drug_unstack_df.loc[grp_drug_unstack_df['Timepoint']==45, 'Placebo'].reset_index(drop='True').get(key=0)
Plavolchange = ((Plavolend - Plavolstart)/Plavolstart)*100



# Create the plot
fig, ax=plt.subplots()

# I set up the height parameter and the xaxis parameter

change=[Capvolchange, Infvolchange, Ketvolchange, Plavolchange]
x_axis=np.arange(len(change))

# define some tick labels

xticks=['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']

#Create function to set the colors of the bars based on the change


def setcolor(change):
    colors=[]
    for i in range(len(change)):
        if change[i]<0:
            colors.append('green')
        else:
            colors.append('red')
    return colors

#store result in colors, which is then used in chart description
colors=(setcolor(change))



# Here I define the barlabels as for formated version of the changes
#  These can then be used to label the change in the plot

barlabels = ["%.0f"% member for member in change]
for i in range(len(barlabels)):
    barlabels[i]=barlabels[i] + '%'


#create the chart

bar=plt.bar(x=x_axis, height=change, color=colors, tick_label=xticks, width=-1, align='edge')

#Use this loop to create a label for each bar that will be added to the chart

for i, v in enumerate(change):
    if change[i]<0:
        ax.text(i-0.6,-6, barlabels[i],color='white')
    else:
        ax.text(i-0.6, 2, barlabels[i], color='white')

#add some additional elements to make the chart more informative and nicer looking        
plt.grid()
plt.ylim(-30,70)
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")

plt.show()
```
